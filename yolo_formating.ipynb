```python
import json
import os

# JSON 파일 경로
json_dir = '/content/drive/MyDrive/산업공학회/산업공학회/yolov5/ｄ/output_json'


# 변환하려는 JSON 파일 이름들 (4개 파일)
json_files = [
    'A02_B02_C00_D01_0811_E08_F07_62_3.json',
    'A02_B02_C00_D01_0811_E07_F07_119_4.json',
    'A02_B02_C00_D01_0811_E08_F07_12_2.json',
    'A02_B02_C00_D01_0811_E07_F07_129_5.json'
]


# 'sample_txt_normal' 폴더 생성
output_dir = '/content/drive/MyDrive/산업공학회/산업공학회/yolov5/ｄ/sample/sample_txt_normal'

os.makedirs(output_dir, exist_ok=True)

# 각 JSON 파일에 대해 변환 작업 수행
for json_file in json_files:
    with open(os.path.join(json_dir, json_file)) as f:
        data = json.load(f)
        items = data['FILE'][0]['ITEMS']
        img_width, img_height = map(int, data['FILE'][0]['RESOLUTION'].split("*"))

        classes = {'불법차량': 0, '정상차량': 1}

        for item in items:
            class_name = item["PACKAGE"]
            class_id = classes[class_name]

            # X1, Y1, X2, Y2 (라벨링구분이 Box인 경우 필수)
            box = item['BOX'].split(',')

            # 박스가 올바른 형식인지 확인
            if len(box) == 4:
                x1, y1, x2, y2 = float(box[0]), float(box[1]), float(box[2]), float(box[3])

                # 바운딩 박스 좌표값을 이용하여 YOLO 형식의 라벨 정보 생성
                dw = 1. / img_width
                dh = 1. / img_height

                x = x1 + x2 / 2.0
                y = y1 + y2 / 2.0
                w = x2
                h = y2

                x_center = x * dw
                width = w * dw
                y_center = y * dh
                height = h * dh

                yolo_label = f"{class_id} {x_center} {y_center} {width} {height}"

                # 파일 이름 생성 (JSON 파일 확장자 대신에 txt 확장자 사용)
                filename = os.path.splitext(json_file)[0] + '.txt'
                txt_file_path = os.path.join(output_dir, filename)

                # 라벨 파일 생성 및 저장
                with open(txt_file_path, 'w') as txt_f:
                    txt_f.write(yolo_label)
            else:
                print(f"파일 '{json_file}'의 BOX 값 형식이 올바르지 않습니다.")
```
