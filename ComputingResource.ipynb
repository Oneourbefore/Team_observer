```python
import matplotlib.pyplot as plt
import numpy as np

fps = np.linspace(10, 30, 100)
stability = (1000 / fps + np.random.normal(0, 2, 100))/10
plt.plot(fps, stability, linestyle='-', color='b')

plt.axvline(x=16, color='r', linestyle='--', label='current model')
plt.axvline(x=24, color='g', linestyle='--', label='existing model')

plt.title('relationship between energy efficiency and FPS')
plt.xlabel('FPS (Frames Per Second)')
plt.ylabel('energy efficiency')

#  
plt.grid(True)

#  
plt.legend()

#  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# FPS  
fps = np.linspace(10, 40, 100)

#    (y = 1/x)
energy_efficiency = 1 / fps

#  
plt.plot(fps, energy_efficiency, linestyle='-', color='b', label='Energy Efficiency')

#  
plt.axvline(x=20, color='r', linestyle='--', label='Optimal FPS')

#    
plt.title('Relationship between Energy Efficiency and FPS')
plt.xlabel('FPS (Frames Per Second)')
plt.ylabel('Energy Efficiency')

#  
plt.grid(True)

#  
plt.legend()

#  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# FPS  
fps = np.linspace(5, 30, 100)

# Computing Resource  ( )
stability_1 = (1000 / fps + np.random.normal(0, 2, 100))/10

# Stability_1   
increased_stability_1 = stability_1[::-1] + 1  #   1 

#  
plt.plot(fps, increased_stability_1)
plt.axvline(x=16, color='r', linestyle='--', label='TO-BE')
plt.axvline(x=24, color='g', linestyle='--', label='AS-IS')

plt.yticks([])
plt.ylabel('')

#     
plt.title('Computing Resource and FPS')
plt.xlabel('FPS (Frames Per Second)')
plt.ylabel('Computing Resource')

#  
plt.grid(True)

#  
plt.legend()

#  
plt.show()
```
