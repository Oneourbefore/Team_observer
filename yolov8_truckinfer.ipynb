```python
import psutil
import matplotlib.pyplot as plt
import time

# CPU 사용량 모니터링
def monitor_cpu(interval):
    cpu_percentages = []
    timestamps = []

    while True:
        cpu_percent = psutil.cpu_percent()
        timestamp = time.time()

        cpu_percentages.append(cpu_percent)
        timestamps.append(timestamp)

        plt.plot(timestamps, cpu_percentages)
        plt.xlabel("Time")
        plt.ylabel("CPU Usage (%)")
        plt.title("CPU Usage Monitor")
        plt.grid(True)
        plt.pause(interval)

# 메모리 사용량 모니터링
def monitor_memory(interval):
    memory_percentages = []
    timestamps = []

    while True:
        memory = psutil.virtual_memory()
        used_memory = memory.used / (1024 * 1024) # 메모리 사용량을 MB 단위로 변환
        available_memory = memory.available / (1024 * 1024) # 사용 가능한 메모리 양을 MB 단위로 변환
        timestamp = time.time()

        memory_percentages.append(used_memory)
        timestamps.append(timestamp)

        plt.plot(timestamps, memory_percentages)
        plt.xlabel("Time")
        plt.ylabel("Memory Usage (MB)")
        plt.title("Memory Usage Monitor")
        plt.grid(True)
        plt.pause(interval)

# 디스크 사용량 모니터링
def monitor_disk(interval):
    disk_percentages = []
    timestamps = []

    while True:
        usage = psutil.disk_usage('/')
        used_disk = usage.used / (1024 * 1024 * 1024) # 디스크 사용량을 GB 단위로 변환
        timestamp = time.time()

        disk_percentages.append(used_disk)
        timestamps.append(timestamp)

        plt.plot(timestamps, disk_percentages)
        plt.xlabel("Time")
        plt.ylabel("Disk Usage (GB)")
        plt.title("Disk Usage Monitor")
        plt.grid(True)
        plt.pause(interval)

# 네트워크 사용량 모니터링
def monitor_network(interval):
    sent_data = []
    recv_data = []
    timestamps = []

    while True:
        network = psutil.net_io_counters()
        sent_bytes = network.bytes_sent / (1024 * 1024) # 송신 데이터 양을 MB 단위로 변환
        recv_bytes = network.bytes_recv / (1024 * 1024) # 수신 데이터 양을 MB 단위로 변환
        timestamp = time.time()

        sent_data.append(sent_bytes)
        recv_data.append(recv_bytes)
        timestamps.append(timestamp)

        plt.plot(timestamps, sent_data, label="Sent")
        plt.plot(timestamps, recv_data, label="Received")
        plt.xlabel("Time")
        plt.ylabel("Network Usage (MB)")
        plt.title("Network Usage Monitor")
        plt.legend()
        plt.grid(True)
        plt.pause(interval)

# 사용 예시
if __name__ == "__main__":
    interval = 1  # 업데이트 간격 (초 단위)
    yolo_command = 'yolo task=detect mode=predict model=/content/drive/MyDrive/ISE_project/09_24_highway/runs/detect/train5/weights/best.pt conf=0.25 source=/content/drive/MyDrive/ISE_project/09_24_highway/pilltest save=True'

    # yolo 명령을 subprocess로 실행
    yolo_process = subprocess.Popen(yolo_command, shell=True)

    # monitor_cpu 호출
    monitor_cpu(interval, yolo_process)
    # monitor_memory(interval)
    # monitor_disk(interval)
    # monitor_network(interval)
```

```python
import subprocess
import psutil
import matplotlib.pyplot as plt
import time

def is_yolo_running():
    try:
        result = subprocess.run(['ps', 'aux'], stdout=subprocess.PIPE)
        output = result.stdout.decode('utf-8')
        if 'yolo' in output:
            return True
        else:
            return False
    except Exception as e:
        print(f"Error checking if YOLO is running: {e}")
        return False

def monitor_cpu(yolo_command):
    cpu_percentages = []

    try:
        # yolo 명령을 subprocess로 실행
        yolo_process = subprocess.Popen(yolo_command, shell=True)

        while is_yolo_running():
            cpu_percent = psutil.cpu_percent()
            cpu_percentages.append(cpu_percent)

            time.sleep(1)

        yolo_process.terminate()  # 객체 탐지가 끝났으면 프로세스 종료

    except Exception as e:
        print(f"Error monitoring CPU usage: {e}")

    return cpu_percentages

if __name__ == "__main__":
    yolo_command1 = 'yolo task=detect mode=predict model=/content/drive/MyDrive/ISE_project/09_24_highway/runs/detect/train5/weights/best.pt conf=0.25 source=/content/drive/MyDrive/ISE_project/09_24_highway/pilltest save=True'

    yolo_command2 = 'yolo task=detect mode=predict model=/content/drive/MyDrive/ISE_project/09_24_highway/runs/detect/train5/weights/best.pt conf=0.25 source=/content/drive/MyDrive/ISE_project/09_24_highway/pilltest2 save=True'

    cpu_percentages1 = monitor_cpu(yolo_command1)  # 첫 번째 객체 탐지는 CPU 사용
    cpu_percentages2 = monitor_cpu(yolo_command2)  # 두 번째 객체 탐지는 CPU 사용

    # 그래프 그리기
    plt.plot(cpu_percentages1, label='16fps (Before Object Detection)')
    plt.plot(cpu_percentages2, label='24fps (Before Object Detection)')
    plt.xlabel("Time")
    plt.ylabel("CPU Usage (%)")
    plt.title("CPU Usage Before Object Detection Comparison")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def calculate_area(cpu_percentages):
    area = 0
    for i in range(1, len(cpu_percentages)):
        x1, x2 = i - 1, i
        y1, y2 = cpu_percentages[i - 1], cpu_percentages[i]
        area += (x2 - x1) * min(y1, y2)
    return area

# 첫 번째 그래프와 두 번째 그래프의 면적 계산
area_24fps = calculate_area(cpu_percentages1)
area_16fps = calculate_area(cpu_percentages2)

print("24fps 일 때 CPU 사용량 그래프 면적 :", area_24fps)
print("16fps 일 때 CPU 사용량 그래프 면적 :", area_16fps)
```

```python
import subprocess
import pynvml
import matplotlib.pyplot as plt
import time

def is_yolo_running():
    try:
        result = subprocess.run(['ps', 'aux'], stdout=subprocess.PIPE)
        output = result.stdout.decode('utf-8')
        if 'yolo' in output:
            return True
        else:
            return False
    except Exception as e:
        print(f"Error checking if YOLO is running: {e}")
        return False

def monitor_gpu(yolo_command):
    pynvml.nvmlInit()
    gpu_handle = pynvml.nvmlDeviceGetHandleByIndex(0)  # Assuming you have only one GPU, change the index if you have multiple GPUs
    gpu_percentages = []

    yolo_process = subprocess.Popen(yolo_command, shell=True)

    while is_yolo_running():
        try:
            gpu_info = pynvml.nvmlDeviceGetUtilizationRates(gpu_handle)
            gpu_percent = gpu_info.gpu
            timestamp = time.time()
            gpu_percentages.append(gpu_percent)

            time.sleep(1)

            if yolo_process.poll() is not None:
                break
        except Exception as e:
            print(f"Error monitoring GPU usage: {e}")
            break
    pynvml.nvmlShutdown()
    return gpu_percentages

if __name__ == "__main__":
    yolo_command1 = 'yolo task=detect mode=predict model=/content/drive/MyDrive/ISE_project/09_24_highway/runs/detect/train5/weights/best.pt conf=0.25 source=/content/drive/MyDrive/ISE_project/09_24_highway/pilltest save=True'

    yolo_command2 = 'yolo task=detect mode=predict model=/content/drive/MyDrive/ISE_project/09_24_highway/runs/detect/train5/weights/best.pt conf=0.25 source=/content/drive/MyDrive/ISE_project/09_24_highway/pilltest2 save=True'

    gpu_percentages1 = monitor_gpu(yolo_command1)
    gpu_percentages2 = monitor_gpu(yolo_command2)

plt.plot(gpu_percentages1, label='24fps')
plt.plot(gpu_percentages2, label='16fps')
plt.xlabel("Time")
plt.ylabel("GPU Usage (%)")
plt.title("GPU Usage Comparison")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import subprocess
import psutil
import matplotlib.pyplot as plt
import time

def is_yolo_running():
    try:
        result = subprocess.run(['ps', 'aux'], stdout=subprocess.PIPE)
        output = result.stdout.decode('utf-8')
        if 'yolo' in output:
            return True
        else:
            return False
    except Exception as e:
        print(f"Error checking if YOLO is running: {e}")
        return False

def record_memory_while_running(yolo_command):
    memory_percentages = []
    timestamps = []

    # yolo 명령을 subprocess로 실행
    yolo_process = subprocess.Popen(yolo_command, shell=True)

    while is_yolo_running():
        memory = psutil.virtual_memory()
        used_memory = memory.used / (1024 * 1024)  # 메모리 사용량을 MB 단위로 변환
        timestamp = time.time()

        memory_percentages.append(used_memory)
        timestamps.append(timestamp)

        time.sleep(1)  # 1초마다 측정

    yolo_process.wait()  # yolo 프로세스가 종료될 때까지 대기

    # 첫 번째 시간을 기준으로 x축 리셋
    timestamps1 = [(t - timestamps[0]) for t in timestamps]
    timestamps2 = [(t - timestamps[0]) for t in timestamps]

    # 첫 번째 그래프
    plt.plot(timestamps1, memory_percentages1, label='24fps')
    plt.xlabel("Time")
    plt.ylabel("Memory Usage (MB)")
    plt.title("Memory Usage of 24fps")
    plt.legend()
    plt.grid(True)
    plt.show()

    # 두 번째 그래프
    plt.plot(timestamps2, memory_percentages2, label='16fps')
    plt.xlabel("Time")
    plt.ylabel("Memory Usage (MB)")
    plt.title("Memory Usage of 16fps")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np

# 각 그래프의 데이터를 numpy 배열로 변환
timestamps1 = np.array(timestamps1)
memory_percentages1 = np.array(memory_percentages1)

timestamps2 = np.array(timestamps2)
memory_percentages2 = np.array(memory_percentages2)

# 공통의 x 축을 생성
common_timestamps = np.linspace(min(timestamps1.min(), timestamps2.min()), max(timestamps1.max(), timestamps2.max()), num=100)

# 두 그래프의 데이터를 새로운 x 축에 대응하도록 보간
interpolated_memory1 = np.interp(common_timestamps, timestamps1, memory_percentages1)
interpolated_memory2 = np.interp(common_timestamps, timestamps2, memory_percentages2)

# 첫 번째 그래프
plt.plot(common_timestamps, interpolated_memory1, label='24fps')
plt.xlabel("Time")
plt.ylabel("Memory Usage (MB)")
plt.title("Memory Usage of 24fps")
plt.legend()
plt.grid(True)

# 두 번째 그래프
plt.plot(common_timestamps, interpolated_memory2, label='16fps')
plt.xlabel("Time")
plt.ylabel("Memory Usage (MB)")
plt.title("Memory Usage of 16fps")
plt.legend()
plt.grid(True)

plt.show()
```
